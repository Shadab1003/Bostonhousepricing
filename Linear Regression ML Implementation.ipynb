```python
## Dividing features into independent and dependent features
X=df.drop('Price',axis=1)
y=df['Price']
```

```python
X.head(1)
```

```python
y.head(1)
```

```python
## Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
## Standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.fit_transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

## Model training

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
## Print the coefficients
print(lr.coef_)
```

```python
# print intercept
print(lr.intercept_)
```

```python
# on which parameters model has been trained
lr.get_params()
```

```python
#prediction with test data
reg_pred=lr.predict(X_test)
reg_pred
```

```python
#3 plot a scatter plot for prediction
plt.scatter(y_test,reg_pred)
```

```python
#Residuals
residuals=y_test-reg_pred
```

```python
#plotting the residuals
sns.kdeplot(residuals)
```

```python
## scatter plot wrt prediction and residuals
#uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

```python
# R squared and adjusted R squared
```

```python
# R squared
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
score
```

```python
# Adjusted R squared
1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
